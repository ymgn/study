```python
from bs4 import BeautifulSoup
```

```python
#1分あたり10リクエストまで
s = scrapelib.Scraper(requests_per_minute=10)
```

```python
r=s.get('http://tabelog.com/osaka/A2706/A270602/R1436/rstLst/?vs=1&sa=%E6%B1%9F%E5%9D%82%E9%A7%85&sk=%25E3%2581%2586%25E3%2581%25A9%25E3%2582%2593&lid=hd_search1&vac_net=&svd=20160731&svt=1900&svps=2&hfc=1&Cat=RC&LstCat=RC01&LstCatD=RC0104&LstCatSD=RC010402&sw=')
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
data = []
```

```python
print(soup)
```

```python
omises = soup.find_all("li",class_="list-rst is-blocklink js-bookmark js-open-newtab")
```

```python
for omise in omises:
    omise_tuika = {}
    namae = omise.find("a",class_="list-rst__rst-name-target cpy-rst-name")
    omise_tuika["namae"] = namae.string
    print(namae.string)
    hyouka = omise.find("span",class_="tb-rating__val tb-rating__val--strong list-rst__rating-val")
    if(hyouka.string != ""):
        omise_tuika["hyouka"] = float(hyouka.string)
    kutikomi = omise.find("em",class_="list-rst__rvw-count-num cpy-review-count")
    omise_tuika["kutikomi"] = int(kutikomi.string)
    data.append(omise_tuika)
```
